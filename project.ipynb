```python
dataB = pd.read_csv("economy-and-growth.csv")
dataB.dropna(axis=0, how='any', inplace=True)
# Has 131 unique countries
clean_dataB_countries = dataB[dataB['Country Name'].isin(filter_valid_countries)]
# print(clean_dataB_countries['Country Name'].nunique())


columns_to_keep = ['Country Name','Country Code', 'Year', 'average_value_GDP (constant 2010 US$)', 'average_value_GDP per capita (constant 2010 US$)','average_value_GDP growth (annual %)']
clean_dataB_countries = clean_dataB_countries[columns_to_keep]
clean_dataB_countries.sort_values(by=['Country Name'], inplace=True)
clean_dataB_countries.sort_values(by=['average_value_GDP (constant 2010 US$)'], inplace=True)

clean_dataB_countries.insert(1, "income_group", 1)

clean_dataB_countries['income_group'] = clean_dataB_countries['Country Code'].map(country_income_dict)

rows_with_nan = clean_dataB_countries[clean_dataB_countries.isna().any(axis=1)]

clean_dataB_countries.dropna(axis=0, how='any', inplace=True)

# clean_dataB_countries.head()

# clean_dataB_countries.head(0)


chart1 = alt.Chart(clean_dataB_countries).mark_point().encode(
    x=alt.X('Year', scale=alt.Scale(domain=[2000, 2020])),
    y='average_value_GDP per capita (constant 2010 US$)',
    color='Country Name:N'
).properties(
    height=300,
    width=300
).interactive()

# chart2 = alt.Chart(clean_dataB_countries).mark_bar().encode(
#     x='Year',
#     y=alt.Y('average_value_GDP growth (annual %)').bin(maxbins=30),
#     color='Country Name:N'
# ).properties(
#     height=300,
#     width=100
# )

chart1

# chart1 | chart2

# clean_dataB_countries

# rows_with_nan['Country Name'].unique()


```
